```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
fruits = pd.read_table('/content/fruit_data_with_colors.txt')
```

```python
fruits.tail(40)
```

```python
lookup_fruit = dict(zip(fruits.fruit_label.unique(),fruits.fruit_name.unique()))
lookup_fruit
```

```python
fruits.shape
```

```python
#Train Test Split
X = fruits.drop(labels='fruit_label',axis=1)
y = fruits['fruit_label']

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
print(X_train.shape)
X_train.head()
```

```python
print(X_test.shape)
X_test.head()
```

```python
print(y_train.shape)
print(y_test.shape)
y_train.head()
```

```python
from matplotlib import cm
cmap = cm.get_cmap('gnuplot')
scatter = pd.scatter_matrix(X_train, c= y_train, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(9,9), cmap=cmap)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X_train['width'], X_train['height'], X_train['color_score'], c = y_train, marker = 'o', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```

```python
#Prediction Model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train, y_train)
```

```python
#accuracy
knn.score(X_test,y_test)
```

```python
fruit_pediction = knn.predict([[100,6.3,8.5]])
lookup_fruit[fruit_pediction[0]]
```
